## Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

## Data Pre-processing

```python
dataset=pd.read_csv('heart_disease_data.csv')
```

```python
dataset.info()
```

1---> Defective Heart
0---> Healthy Heart

```python
dataset.groupby('target').mean()
```

```python
X=dataset.drop(columns='target', axis=1)
Y=dataset['target']
```

```python
print(X)
```

```python
print(Y)
```

## Training and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
X_train_prediction=model.predict(X_train)
train_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(f"Training Accuracy is :{train_accuracy}")
```

```python
X_test_prediction=model.predict(X_test)
test_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(f"Testing Accuracy is :{test_accuracy}")
```

## Making a predictive system

```python
input_data=(55,1,0,140,217,0,1,111,1,5.6,0,0,3)
input_data_as_nparray=np.asarray(input_data)
reshaped_data=input_data_as_nparray.reshape(1,-1)
prediction=model.predict(reshaped_data)

print(prediction)
if (prediction[0]==0):
    print("Healthy Heart")
else:
    print("Defective Heart")
```
