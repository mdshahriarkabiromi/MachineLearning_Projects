## Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression 
```

## Seperating the data for analysis

```python
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

## Under Sampling

As our dataset is highly unbalanced. We need to balance it. We got 284315 legit data and 492 fraud data. We need to create a dataset which will have balance distribution

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns="Class",axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Training and Testing Dataset

```python
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

## Model Training

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
accuracy_traindata=model.predict(X_train)
training_accuracy=accuracy_score(accuracy_traindata,Y_train)
print(f"The training accuracy is :{training_accuracy}")
```

```python
accuracy_testdata=model.predict(X_test)
testing_accuracy=accuracy_score(accuracy_testdata,Y_test)
print(f"The testing accuracy is :{testing_accuracy}")
```
